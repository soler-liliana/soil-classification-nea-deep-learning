**SeaLake:** Grandes cuerpos de agua estáticos, como mares, lagos y embalses.

ResNet 34 con Cross Validation

```python
import torch
import torchvision
import torchvision.models as models
import torchvision.transforms as transforms
import torch.nn as nn
import torch.optim as optim
import os
from torch.utils.data import DataLoader, random_split, Subset
from torchvision.datasets import ImageFolder
import matplotlib.pyplot as plt
import numpy as np
import random
from sklearn.model_selection import KFold
from torch.optim.lr_scheduler import ReduceLROnPlateau

# Semillas para reproducibilidad
seed = 42
random.seed(seed)
np.random.seed(seed)
torch.manual_seed(seed)
torch.cuda.manual_seed(seed)
torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False

# Transformaciones para entrenamiento y validación/test
train_transforms = transforms.Compose([
    transforms.Resize((224, 224)),  # Cambia el tamaño de las imágenes a 224x224 píxeles. 
    transforms.RandomRotation(40),  # Aplica una rotación aleatoria de hasta 40 grados.
    transforms.RandomHorizontalFlip(),  # Invierte la imagen horizontalmente en forma aleatoria.
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),  # Modifica brillo, contraste y color
    transforms.RandomApply([transforms.GaussianBlur(kernel_size=3)], p=0.2),  # Aplica desenfoque con 20% de probabilidad
    transforms.ToTensor(),  # Convierte la imagen en un tensor.
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])  
])

val_transforms = transforms.Compose([
    transforms.Resize((224, 224)),  # Cambia el tamaño de las imágenes a 224x224.
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Directorio de datos
data_dir = '/home1/prog/soler/Trabajo_Final/EuroSAT_RGB'

# Cargar el dataset completo
full_dataset = ImageFolder(data_dir, transform=train_transforms)

# Obtener los nombres de las clases
class_names = full_dataset.classes
print("Clases:", class_names)

# Dividir en Train+Val (80%) y Test (20%)
train_val_size = int(0.8 * len(full_dataset))
test_size = len(full_dataset) - train_val_size
train_val_dataset, test_dataset = random_split(full_dataset, [train_val_size, test_size])

# Separar Train (80%) y Validation (20%) dentro de Train+Val
train_size = int(0.8 * train_val_size)  # 80% de Train+Val para Train
val_size = train_val_size - train_size  # 20% de Train+Val para Validación
train_dataset, val_dataset = random_split(train_val_dataset, [train_size, val_size])

# Aplicar transformaciones específicas para validación y prueba
test_dataset.dataset.transform = val_transforms

# Número de folds para K-Fold Cross Validation
k_folds = 5
kfold = KFold(n_splits=k_folds, shuffle=True, random_state=seed)

# Función para crear el modelo
def create_model():
    model = models.resnet34(pretrained=True)
    num_features = model.fc.in_features
    model.fc = nn.Linear(num_features, 10)  # 10 clases
    return model.to(device)

# Función para entrenar un solo fold
def train_one_fold(model, train_loader, val_loader, criterion, optimizer, scheduler, epochs=10):
    train_losses = []
    val_losses = []
    train_accuracies = []
    val_accuracies = []

    for epoch in range(epochs):
        model.train()
        train_loss = 0.0
        correct = 0

        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            train_loss += loss.item()
            preds = torch.argmax(outputs, dim=1)
            correct += (preds == labels).sum().item()

        train_loss /= len(train_loader)
        train_acc = correct / len(train_loader.dataset)
        train_losses.append(train_loss)
        train_accuracies.append(train_acc)

        # Validación
        model.eval()
        val_loss = 0.0
        correct = 0

        with torch.no_grad():
            for images, labels in val_loader:
                images, labels = images.to(device), labels.to(device)
                outputs = model(images)
                loss = criterion(outputs, labels)
                val_loss += loss.item()
                preds = torch.argmax(outputs, dim=1)
                correct += (preds == labels).sum().item()

        val_loss /= len(val_loader)
        val_acc = correct / len(val_loader.dataset)
        val_losses.append(val_loss)
        val_accuracies.append(val_acc)

        print(f"Epoch [{epoch+1}/{epochs}], Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f}, "
              f"Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.4f}")

        scheduler.step(val_loss)  # Ajustar el learning rate

    return train_losses, val_losses, train_accuracies, val_accuracies

# Función para validación cruzada
def cross_validate(dataset, kfold, epochs=10, batch_size=32):
    metrics_per_fold = []

    for fold, (train_idx, val_idx) in enumerate(kfold.split(dataset)):
        print(f"\n===== Fold {fold+1}/{k_folds} =====")

        # Crear subconjuntos para train y val
        train_subset = Subset(dataset, train_idx)
        val_subset = Subset(dataset, val_idx)

        # Asignar transformaciones
        train_subset.dataset.transform = train_transforms
        val_subset.dataset.transform = val_transforms

        # DataLoaders
        train_loader = DataLoader(train_subset, batch_size=batch_size, shuffle=True)
        val_loader = DataLoader(val_subset, batch_size=batch_size, shuffle=False)

        # Crear y entrenar el modelo
        model = create_model()
        criterion = nn.CrossEntropyLoss()
        optimizer = torch.optim.SGD(model.parameters(), lr=5e-4, momentum=0.9, weight_decay=5e-3)
        scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.1, patience=3, verbose=True)

        train_losses, val_losses, train_accuracies, val_accuracies = train_one_fold(
            model, train_loader, val_loader, criterion, optimizer, scheduler, epochs)

        metrics_per_fold.append({
            'train_loss': train_losses,
            'val_loss': val_losses,
            'train_acc': train_accuracies,
            'val_acc': val_accuracies,
            'fold': fold + 1
        })

    return metrics_per_fold

# Función para graficar métricas
def plot_metrics(metrics_per_fold):
    for metric in ['train_loss', 'val_loss', 'train_acc', 'val_acc']:
        plt.figure(figsize=(10, 5))
        for fold_metrics in metrics_per_fold:
            plt.plot(fold_metrics[metric], label=f'Fold {fold_metrics["fold"]}')
        plt.xlabel("Epoch")
        plt.ylabel(metric.replace('_', ' ').title())
        plt.legend()
        plt.title(f"{metric.replace('_', ' ').title()} por Fold")
        plt.show()

# Ejecutar validación cruzada
epochs = 10
batch_size = 32
metrics_per_fold = cross_validate(train_val_dataset, kfold, epochs, batch_size)
plot_metrics(metrics_per_fold)

# Entrenar modelo final con todos los datos de entrenamiento + validación
def train_final_model(train_val_dataset, epochs=10, batch_size=32):
    train_val_dataset.dataset.transform = train_transforms
    train_loader = DataLoader(train_val_dataset, batch_size=batch_size, shuffle=True)

    model = create_model()
    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.SGD(model.parameters(), lr=5e-4, momentum=0.9, weight_decay=5e-3)
    scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.1, patience=3, verbose=True)

    for epoch in range(epochs):
        model.train()
        train_loss = 0.0
        correct = 0

        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            train_loss += loss.item()
            preds = torch.argmax(outputs, dim=1)
            correct += (preds == labels).sum().item()

        train_loss /= len(train_loader)
        train_acc = correct / len(train_loader.dataset)

        print(f"Epoch [{epoch+1}/{epochs}], Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f}")

        scheduler.step(train_loss)

    return model

# Evaluar el modelo final en el conjunto de prueba
def evaluate_model(model, test_loader):
    model.eval()
    correct = 0
    total = 0
    test_loss = 0.0
    criterion = nn.CrossEntropyLoss()

    with torch.no_grad():
        for images, labels in test_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            loss = criterion(outputs, labels)
            test_loss += loss.item()
            preds = torch.argmax(outputs, dim=1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)

    test_loss /= len(test_loader)
    test_acc = correct / total
    print(f"Test Loss: {test_loss:.4f}, Test Accuracy: {test_acc:.4f}")

# Entrenar y evaluar el modelo final
final_model = train_final_model(train_val_dataset, epochs=25, batch_size=32)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
evaluate_model(final_model, test_loader)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns

def evaluate_with_metrics(model, test_loader, class_names):
    model.eval()
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for images, labels in test_loader:
            images = images.to(device)
            outputs = model(images)
            preds = torch.argmax(outputs, dim=1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.numpy())

    # Matriz de confusión con seaborn
    cm = confusion_matrix(all_labels, all_preds)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
    plt.xlabel('Predicción')
    plt.ylabel('Etiqueta Verdadera')
    plt.title('Matriz de Confusión')
    plt.xticks(rotation=90)
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()

    # Reporte de métricas
    report = classification_report(all_labels, all_preds, target_names=class_names, digits=4)
    print("Reporte de Clasificación:\n")
    print(report)

# Llamada a la función
evaluate_with_metrics(final_model, test_loader, class_names)
```

```python
import torchvision.datasets as datasets
from PIL import Image
from time import time

def main():
    # Configuración inicial
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"\n🚀 Dispositivo de ejecución: {device}")
    
    # 1. Carga del modelo guardado
    print("\n🔍 Cargando modelo...")
    model = torch.load("modelo_eurosat_completo_1.pth", map_location=device)
    model.eval()
    model.to(device)
    print("Modelo cargado correctamente")

    # 2. Preparación de datos
    print("\n Preparando datos de prueba...")
    image_dir = "./Images_Ctes"
    
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.Lambda(lambda x: x.convert('RGB') if x.mode != 'RGB' else x),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])
    
    test_dataset = datasets.ImageFolder(root=image_dir, transform=transform)
    class_labels = test_dataset.classes
    print(f" Clases detectadas ({len(class_labels)}): {class_labels}")
    
    test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
    image_paths = [sample[0] for sample in test_dataset.samples]

    # 3. Evaluación del modelo
    print("\n Evaluando modelo...")
    predictions = []
    true_labels = []
    probabilities = []
    
    start_time = time()
    with torch.no_grad():
        for batch, targets in test_loader:
            batch = batch.to(device)
            outputs = model(batch)
            
            # Guardar probabilidades para análisis posterior
            probs = torch.nn.functional.softmax(outputs, dim=1)
            probabilities.extend(probs.cpu().numpy())
            
            preds = torch.argmax(outputs, dim=1)
            predictions.extend(preds.cpu().numpy())
            true_labels.extend(targets.cpu().numpy())
    
    # Convertir índices a nombres de clase
    pred_labels = [class_labels[p] for p in predictions]
    true_labels_names = [class_labels[t] for t in true_labels]
    
    print(f" Tiempo de evaluación: {time() - start_time:.2f} segundos")

    # 4. Métricas de evaluación
    print("\n Métricas de evaluación:")
    print(classification_report(true_labels_names, pred_labels, target_names=class_labels))
    
    # Matriz de confusión
    cm = confusion_matrix(true_labels_names, pred_labels, labels=class_labels)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=class_labels, yticklabels=class_labels)
    plt.title('Matriz de Confusión')
    plt.xlabel('Predicciones')
    plt.ylabel('Verdaderos')
    plt.xticks(rotation=90)
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()

    # 5. Visualización de ejemplos
    print("\n Visualizando ejemplos...")
    num_samples = min(5, len(image_paths))
    sample_indices = random.sample(range(len(image_paths)), num_samples)
    
    fig, axes = plt.subplots(1, num_samples, figsize=(15, 5))
    if num_samples == 1:
        axes = [axes]
    
    for i, idx in enumerate(sample_indices):
        img = Image.open(image_paths[idx])
        axes[i].imshow(img)
        
        # Mostrar probabilidad de la clase predicha
        pred_prob = probabilities[idx][predictions[idx]]
        axes[i].set_title(f"Real: {true_labels_names[idx]}\nPred: {pred_labels[idx]}\nConf: {pred_prob:.2f}")
        axes[i].axis('off')
    
    plt.tight_layout()
    plt.show()

    # 6. Guardado de resultados
    print("\n Guardando resultados...")
    results_df = pd.DataFrame({
        "Imagen": image_paths,
        "Etiqueta Real": true_labels_names,
        "Predicción": pred_labels,
        "Probabilidad": [max(prob) for prob in probabilities],
        **{f"Prob_{cls}": [prob[i] for prob in probabilities] 
           for i, cls in enumerate(class_labels)}
    })
    
    results_df.to_csv("resultados_clasificacion_detallados_1.csv", index=False)
    print(" Resultados guardados en 'resultados_clasificacion_detallados_1.csv'")

if __name__ == "__main__":
    main()
```
